```python
print("== CLI 게시판 실행 ==")

last_article_id = 0
articles = [] # 게시글 저장소

while True:
    cmd = input("명령어 ) ").strip()

    if cmd == "exit":
      break

    elif cmd == "c write":
      last_article_id += 1
      id = last_article_id
      title = input("제목 : ")
      body = input("내용 : ")
      article = {"id" : id, "title" : title, "body" : body}
      # print(article)
      articles.append(article)
      print("{}번 글이 생성되었습니다".format(last_article_id))

    elif cmd == "article list":
      if not articles:
        print("게시글이 없습니다")
      else:
        print("=========================================")
        print("   번호      /     제목       /     내용   ")
        for a in reversed(articles):
          print("   {}      /     {}       /     {}   ".format(a["id"], a["title"], a["body"]))
        print("=========================================")
    elif cmd.startswith("article delete"):
      cmdBits = cmd.split(" ")

      if len(cmdBits) < 3:
        print("명령어를 확인해주세요(길이 모자람)")
        continue

      if not cmdBits[2].isdigit():
        print("명령어를 확인해주세요(번호대신 str 입력)")
        continue

      deletedId = int(cmdBits[2])

      found = False # 못찾았다 를 가정

      for article in articles:
        if article["id"] == deletedId:
          articles.remove(article)
          print("{}번 글이 삭제되었습니다".format(cmdBits[2]))
          found = True # 진짜 찾았다!
          break
      if found == False: # 순회해서 다 뒤져봤는데 아직 found가 False라는 거는 True가 된 적이 없었다
      # else: # for ~ else 표현도 있음
        print("{}번 글은 없습니다".format(cmdBits[2]))
    elif cmd.startswith("article edit"):
      cmdBits = cmd.split(" ")

      if len(cmdBits) < 3:
        print("명령어를 확인해주세요(길이 모자람)")
        continue

      if not cmdBits[2].isdigit():
        print("명령어를 확인해주세요(번호대신 str 입력)")
        continue

      editedId = int(cmdBits[2])

      found = False


    elif cmd.startswith("article edit"):
      cmdBits = cmd.split(" ")

      for article in articles:
        if article["id"] == editedId:

         new_title = input("새로운 title 입력: ")
         new_body = input("새로운 body 입력: ")

         article.update("title"= new_title)
         article.update("body"= new_body)

          # cmd.update(input());
          #articles.update({"title": new_title,"body": new_body}):
          #print(1)
          #new_title = input("제목 : ")
          #print(2)
          #new_body = input("내용 : ")

          #update_data = {}
          #if new_name: update_data['name'] = new_name
          #if new_age: update_data['age'] = int(new_age)
         print("{}번 글이 수정되었습니다".format(int(cmdBits[2])))
          for a in reversed(articles):
            print("   {}      /     {}       /     {}   ".format(a["id"], a["title"], a["body"]))


    else:
      print("지원하지 않는 기능입니다")

print("== CLI 게시판 종료 ==")
```
