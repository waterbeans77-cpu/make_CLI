```python
from datetime import datetime

print("== CLI 게시판 실행 ==")

last_article_id = 1
articles = [{"id" : 1, "regDate" : "2025-12-12 12:12:12", "updateDate": "2025-12-12 12:12:12", "title" : "제목1", "body" : "내용1"}, ] # 게시글 저장소

while True:
    cmd = input("명령어 ) ").strip()

    if cmd == "exit":
      break

    elif cmd == "article write":
      last_article_id += 1
      id = last_article_id
      title = input("제목 : ")
      body = input("내용 : ")
      regDate = datetime.now().strftime("%Y-%m-%d %H:%M:%S")
      updateDate = regDate
      article = {"id" : id, "regDate" : regDate, "updateDate": updateDate, "title" : title, "body" : body}
      articles.append(article)
      print("{}번 글이 생성되었습니다".format(last_article_id))

    elif cmd == "article list":
      if not articles:
        print("게시글이 없습니다")
      else:
        today = datetime.now().strftime("%Y-%m-%d %H:%M:%S")
        print("=========================================")
        print("   번호      /     제목       /     내용   /     작성일   ")
        for a in reversed(articles):
          # 비교 대상 주의
          if today.split(" ")[0] != a["regDate"].split(" ")[0]:
            print("   {}      /     {}       /     {}          /     {}   ".format(a["id"], a["title"], a["body"],a["regDate"].split(" ")[0]))
          if today.split(" ")[0] == a["regDate"].split(" ")[0]:
            print("   {}      /     {}       /     {}          /     {}   ".format(a["id"], a["title"], a["body"],a["regDate"].split(" ")[1]))
        print("=========================================")
    elif cmd.startswith("article delete"):
      cmdBits = cmd.split(" ")

      if len(cmdBits) < 3:
        print("명령어를 확인해주세요(길이 모자람)")
        continue

      if not cmdBits[2].isdigit():
        print("명령어를 확인해주세요(번호대신 str 입력)")
        continue

      deletedId = int(cmdBits[2])

      found = False # 못찾았다 를 가정

      for article in articles:
        if article["id"] == deletedId:
          articles.remove(article)
          print("{}번 글이 삭제되었습니다".format(cmdBits[2]))
          found = True # 진짜 찾았다!
          break
      if found == False: # 순회해서 다 뒤져봤는데 아직 found가 False라는 거는 True가 된 적이 없었다
      # else: # for ~ else 표현도 있음
        print("{}번 글은 없습니다".format(cmdBits[2]))
    elif cmd.startswith("article edit"):
      cmdBits = cmd.split(" ")

      if len(cmdBits) < 3:
        print("명령어를 확인해주세요(길이 모자람)")
        continue

      if not cmdBits[2].isdigit():
        print("명령어를 확인해주세요(번호대신 str 입력)")
        continue

      editedId = int(cmdBits[2])

      found = False

      for article in articles:
        if article["id"] == editedId:
          # 기존 제목 내용 보여주기
          print("기존 제목 : {}".format(article["title"]))
          print("기존 내용 : {}".format(article["body"]))
          # 새로 제목 내용 입력받아서 기존 내용 업데이트
          article["title"] = input("새 제목 : ")
          article["body"] = input("새 내용 : ")
          article["updateDate"] = datetime.now().strftime("%Y-%m-%d %H:%M:%S")
          print("{}번 글이 수정되었습니다".format(cmdBits[2]))
          found = True
          break
      if found == False:
        print("{}번 글은 없습니다".format(cmdBits[2]))
    elif cmd.startswith("article detail"):
      cmdBits = cmd.split(" ")

      if len(cmdBits) < 3:
        print("명령어를 확인해주세요(길이 모자람)")
        continue

      if not cmdBits[2].isdigit():
        print("명령어를 확인해주세요(번호대신 str 입력)")
        continue

      detailId = int(cmdBits[2])

      found = False

      for article in articles:
        if article["id"] == detailId:
          print("번호 : {}".format(article["id"]))
          print("작성 날짜 : {}".format(article["regDate"]))
          print("수정 날짜 : {}".format(article["updateDate"]))
          print("제목 : {}".format(article["title"]))
          print("내용 : {}".format(article["body"]))
          found = True
          break
      if found == False:
        print("{}번 글은 없습니다".format(cmdBits[2]))
    else:
      print("지원하지 않는 기능입니다")

print("== CLI 게시판 종료 ==")

```
